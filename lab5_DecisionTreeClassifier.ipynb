```python
#IMPORT ALL ESSENTIAL LIBRARIES 
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from six import StringIO
from sklearn.datasets import load_diabetes
from sklearn import tree

from IPython.display import Image  
import pydotplus
```

```python
#LOAD DATA from the DATASET
data = pd.read_csv(r'/content/sample_data/diabetes.csv')
#Shape of the dataset i.e; rows and columns
print("Dataset Shape:",data.shape)
#Displaying the data from the dataset 
data.head()
```

```python
X = data.iloc[:, :8]
Y = data.iloc[:, 8]
X = np.array(X)
Y = np.array(Y)
```

```python
X
```

```python
Y
```

```python
tree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
tree.fit(X, Y)
```

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png(), width =1000, height = 1000)
```

```python
predict = tree.predict(X)
predict
```

```python
from sklearn import metrics
print('Accuracy: ', metrics.accuracy_score(Y, predict))
print('Confusion Matrix\n', metrics.confusion_matrix(Y, predict))
```
