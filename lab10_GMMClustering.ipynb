```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
```

```python
mobile_data=pd.read_csv(r'/content/sample_data/mobile_price.csv')
mobile_data.head(20)
```

```python
mobile_data.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x=scaler.fit_transform(mobile_data)
x
```

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
train, test = train_test_split(x, train_size=0.7, random_state=10)
train
```

```python
from sklearn.mixture import GaussianMixture
gm=GaussianMixture(n_components=2, random_state=0).fit(train)
GG=gm.predict(train)
GG
```

```python
from sklearn.metrics import silhouette_score
accuracy=silhouette_score(train, GG)
accuracy
```

**GAUSSIAN MIXTURE MODEL CLUSTERING ATTEMPT**

```python
from sklearn.mixture import GaussianMixture
gm=GaussianMixture(n_components=10, random_state=0).fit(train)
GG=gm.predict(train)
GG
```

```python
from sklearn.metrics import silhouette_score
accuracy=silhouette_score(train, GG)
accuracy
```
